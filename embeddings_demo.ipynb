```python
from sentence_transformers import SentenceTransformer, util
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define 10 sentences across Cricket, Cooking, and Cybersecurity
sentences = [
    # Topic: Cricket
    "The batsman scored a brilliant century in the final match of the tournament.",
    "A fast yorker shattered the stumps, dismissing the dangerous opening batter.",
    "The spin bowler delivered a tight spell, conceding very few runs.",
    "The fielding team dropped a crucial catch right at the boundary line.",

    # Topic: Cooking
    "Add a pinch of salt and simmer the tomato sauce for twenty minutes.",
    "Whisk the eggs and sugar vigorously until the mixture becomes light and fluffy.",
    "Bake the chocolate cake in a preheated oven at 350 degrees Fahrenheit.",

    # Topic: Cybersecurity
    "The enterprise firewall successfully blocked a massive distributed denial of service attack.",
    "Always use multi-factor authentication to secure your online banking accounts.",
    "Malicious hackers exploited a dangerous zero-day vulnerability in the web server."
]

print("Sentences loaded successfully.")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(sentences)
```

```python
# Define the new query sentence
query = "The bowler took three wickets in one over"

# Generate embedding for the query
query_embedding = model.encode(query)

# Compute cosine similarities between the query and all 10 sentences
# We extract the first row [0] since we only have one query
query_scores = util.cos_sim(query_embedding, embeddings)[0].numpy()

# Get the indices of the top 2 highest scores
# argsort sorts ascending, so we reverse it with [::-1] and take the first 2
top_2_indices = np.argsort(query_scores)[::-1][:2]

print(f"Query: '{query}'\n")
print("Top 2 Most Similar Sentences:\n")

for rank, idx in enumerate(top_2_indices, start=1):
    print(f"Rank {rank}:")
    print(f"Sentence: {sentences[idx]}")
    print(f"Similarity Score: {query_scores[idx]:.4f}\n")
```
